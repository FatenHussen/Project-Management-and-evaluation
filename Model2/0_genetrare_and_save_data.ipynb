```python
import pandas as pd
import numpy as np
import random
```

```python
# Define the dimensions of the dataset
num_rows = 2500
num_features = 12
min_values_per_feature = 5
max_values_per_feature = 16
columns=["فكرة المشروع" , "دراسة مرجعية" , "دراسة تحليلية" , "تصميم" , "برمجة" , "إدارة" , "GITHUB" , "اختبارات" , "النتائج" , "القيمة المضافة" , "التقرير" , "العرض"]
```

```python
dataset1 = []
num_rows1=650

for i in range(num_rows1):
    row = []
    for j in range(num_features):
        if (j == 0):
            values = random.choice([6, 8])
        elif (j == 3) or (j == 4) :
            values = random.randint(5, 10)
        else:
            values = random.randint(4, 8)
        row.append(values)
    dataset1.append(row)
```

```python
dataset2 = []
num_rows2=650

for i in range(num_rows2):
    row = []
    for j in range(num_features):
        if (j == 0):
            values = random.choice([6, 8])
        elif (j == 3) or (j == 4) :
            values = random.randint(0, 5)
        else:
            values = random.randint(0, 4)
        row.append(values)
    dataset2.append(row)
```

```python
dataset3 = []
num_rows3=1200

for i in range(num_rows3):
    row = []
    for j in range(num_features):
        if (j == 0):
            values = random.choice([6, 8])
        elif (j == 3) or (j == 4) :
            values = random.randint(0, 10)
        else:
            values = random.randint(0, 8)
        row.append(values)
    dataset3.append(row)
```

```python
dataset= dataset1 + dataset2 + dataset3
```

```python
df = pd.DataFrame(dataset, columns=columns)
df
```
